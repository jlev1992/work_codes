```python
import os
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
rc('font',**{'family':'serif','serif':['Times']})
rc('text', usetex=True)
from matplotlib.pyplot import figure
figure(num=None, figsize=(16, 12), dpi=80, facecolor='w', edgecolor='k')

from itertools import groupby
import numpy as np

num_bands=200
num_k=421
E_fermi=0.0 #16.1610

band_dat=np.zeros((num_k,num_bands+1))

f1=open('nbas_fr_bands_gnu')
k=0
i=1
for line in f1:
    line = line.split()
    if len(line)>0:
        band_dat[k,0]=float(line[0])
        band_dat[k,i]=float(line[1])-E_fermi
        k=k+1
    if len(line)==0:
        i=i+1
        k=0

print(band_dat)

hsp=[0.0,0.9972,1.5626,2.3988,3.3548,4.3108,5.6073,np.max(band_dat[:,0])]
plt.xticks((hsp),('$\Gamma$','$\Sigma_{0}$','$\Sigma$','Z','N','$\Gamma$','X','P'),fontsize=32)
plt.yticks(fontsize=32)
plt.ylim(5,20)
plt.xlim((0,np.max(band_dat[:,0])))
plt.ylabel('Energy (eV)',fontsize=32)
plt.gca().xaxis.grid(True)

for i in range(1,num_bands-1):
    plt.plot(band_dat[:,0],band_dat[:,i],color='blue',linewidth=3)

    
fermi_linx = [0,np.max(band_dat[:,0])]
fermi_liny = [0,0]
plt.plot(fermi_linx,fermi_liny,'--',color='grey', linewidth=1)
plt.savefig('nbas_bands_FR')

#tCBAB_new path
#hsp=[0.0,0.7145,2.3057,3.4884,4.1030,np.max(band_dat[:,0])]
#plt.xticks((hsp),('Z','$\Gamma$','R','X','$\Gamma$','V'),fontsize=32)
```
